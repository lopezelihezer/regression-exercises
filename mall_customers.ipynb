```python
quant_vars = ["annual_income", "spending_score", "age"]
cat_vars = ["gender"]
target = "spending_score"
```

```python
def plot_variable_pairs(df):
    for col in df.columns:
        print(f'{col}(x) vs {target}(y)')
        x = df[col]
        y = df[target]
        plt.plot(x, y, 'o')
        m, b = np.polyfit(x, y, 1)
        plt.plot(x, m*x + b)
        plt.show()
        print('__________________')
    return print('done')
```

```python
plot_variable_pairs(df)
```
